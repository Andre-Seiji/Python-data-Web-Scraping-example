```python
# instalando pacote (kora)
# kora -> diversas aplicações para usar o python. Uma dessas é o selenium
!pip install kora -q

from bs4 import BeautifulSoup
import pandas as pd
from kora.selenium import wd
import numpy as np
```

```python
#acessando usando o Selenium
url = 'https://worldometers.info/coronavirus/'
wd.get(url)

#Verificando o conteúdo da página
conteudo=wd.page_source

#Desde o início do conteúdo
conteudo[:500]

#Criando o BeautifulSoup
soup=BeautifulSoup(conteudo, 'html.parser')

#Criando a tabela
tabela=soup.find('table')

#Encontrar o elemento do cookie
elem_cookie=wd.find_elements_by_class_name('cc-compliance')
elem_cookie[0].click()

```

```python
def processo(matriz):
  matriz.fillna(0, inplace = True)
  matriz=(matriz.replace('[\+\,]',"", regex = True))

  matriz['NewCases']=matriz['NewCases'].astype(int)
  matriz['NewDeaths']=matriz['NewDeaths'].astype(int)
  matriz['NewRecovered']=matriz['NewRecovered'].astype(int)
  matriz['TotalRecovered']=matriz['TotalRecovered'].astype(int)
  matriz['ActiveCases']=matriz['ActiveCases'].astype(int)

  #Referência para comparar valor total com o da página
  ref=matriz.iloc[:1,]
  ref

  #Cabeçalho
  nomes=(matriz.columns.values)

  #Só países
  paises=matriz.iloc[1:]
  
  for i in range(1,14):
    print(nomes[i])
 
    if paises[nomes[i]].sum(axis=0)==ref[nomes[i]].sum(axis=0):
      print('Correto!\n\n')
        
    else:
      print('Valor não compatível\n\n')

  print('##### Obs: Os valores "Tot Cases/1M pop", "Deaths/1M pop", "Total Tests", "Tests/1M pop" e \n"Population" não possuem valores totais na tabela original\n\n')
  exp=matriz.iloc[:1,:9]

  exp
  exp.to_excel('total.xlsx', sheet_name='Test', na_rep='#N/A', header=True, index=False)

#Obter os dados da Ásia
#Encontrar o elemento na página
elemento_atual = wd.find_elements_by_partial_link_text('Asia')
elemento_atual[0].click()

#Obtendo o conteúdo da página HTML após o clique
conteudo_novo=wd.page_source

#Fazendo o Parse dos dados do país
soup=BeautifulSoup(conteudo_novo, 'html.parser')
tabela_atual=soup.find(name='table')

df_atual=pd.read_html(str(tabela_atual), index_col=0)[0]

matriz=df_atual.head(50)
matriz
for i in range(1,2):
  print('Ásia\n')
  processo(matriz)
```
